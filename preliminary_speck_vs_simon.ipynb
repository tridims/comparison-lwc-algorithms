### Pengujian Performa Preliminary - Waktu Eksekusi

```python
from simon import SimonCipher
from speck import SpeckCipher
import time
import pandas as pd

# Block modes to test
block_modes = ['ECB', 'CTR', 'CBC', 'PCBC', 'CFB', 'OFB']

# Initialize a dictionary to store the results
results = {}
key_data = _key_data.copy()

# Iterate over the test data
for plaintext in test_data:
    results[plaintext] = {}
    key_int = key_data.pop()

    # Convert the plaintext to an integer
    plaintext_int = sum([ord(c) << (8 * i) for i, c in enumerate(plaintext)])

    # Test Speck cipher
    for mode in block_modes:
        speck_cipher = SpeckCipher(key_int, mode=mode)

        # Measure encryption time
        start_time = time.time()
        ciphertext = speck_cipher.encrypt(plaintext_int)
        end_time = time.time()

        # Calculate encryption time in milliseconds
        encryption_time_ms = (end_time - start_time) * 1000

        # Store the result in the dictionary
        results[plaintext]["Key"] = key_int
        results[plaintext][f"Speck_{mode}"] = encryption_time_ms

    # Test Simon cipher
    for mode in block_modes:
        simon_cipher = SimonCipher(key_int, mode=mode)

        # Measure encryption time
        start_time = time.time()
        ciphertext = simon_cipher.encrypt(plaintext_int)
        end_time = time.time()

        # Calculate encryption time in milliseconds
        encryption_time_ms = (end_time - start_time) * 1000

        # Store the result in the dictionary
        results[plaintext]["Key"] = key_int
        results[plaintext][f"Simon_{mode}"] = encryption_time_ms

# Create a pandas DataFrame from the results
df = pd.DataFrame.from_dict(results, orient="index")

# Save the DataFrame to a CSV file
df.to_csv("encryption_times.csv", index_label="Plaintext")
```

```python
def calculate_avalanche_effect(ciphertext1, ciphertext2):
    ciphertext1_bin = bin(ciphertext1)[2:]
    ciphertext2_bin = bin(ciphertext2)[2:]

    # Pad the shorter binary string with zeros if the lengths are unequal
    max_len = max(len(ciphertext1_bin), len(ciphertext2_bin))
    ciphertext1_bin = ciphertext1_bin.zfill(max_len)
    ciphertext2_bin = ciphertext2_bin.zfill(max_len)

    diff_bits = sum(bit1 != bit2 for bit1, bit2 in zip(ciphertext1_bin, ciphertext2_bin))
    total_bits = len(ciphertext1_bin)
    avalanche_effect = diff_bits / total_bits
    return avalanche_effect
```

```python

from simon import SimonCipher
from speck import SpeckCipher
import pandas as pd

def string_to_int(plaintext):
    return sum([ord(c) << (8 * i) for i, c in enumerate(plaintext)])

# Block modes to test
block_modes = ['CTR', 'CBC', 'PCBC', 'CFB', 'OFB']

# Initialize a dictionary to store the results
results = {}

key_data = _key_data.copy()

# Iterate over the test data and modified data
for plaintext, modified_plaintext in zip(test_data, modified_data):
    results[plaintext] = {}
    key_int = key_data.pop()
    
    # Convert the plaintexts to integers
    plaintext_int = string_to_int(plaintext)
    modified_plaintext_int = string_to_int(modified_plaintext)
    
    # Test Speck cipher
    for mode in block_modes:
        speck_cipher = SpeckCipher(key_int, mode=mode)
        
        ciphertext = speck_cipher.encrypt(plaintext_int)
        modified_ciphertext = speck_cipher.encrypt(modified_plaintext_int)
        
        # Calculate avalanche effect
        avalanche_effect = calculate_avalanche_effect(ciphertext, modified_ciphertext)
        
        # Store the results in the dictionary
        results[plaintext]["Modified_Plaintext"] = modified_plaintext
        results[plaintext]["Key"] = key_int
        results[plaintext][f"Speck_{mode}_Avalanche"] = avalanche_effect
    
    # Test Simon cipher
    for mode in block_modes:
        simon_cipher = SimonCipher(key_int, mode=mode)
        
        ciphertext = simon_cipher.encrypt(plaintext_int)
        modified_ciphertext = simon_cipher.encrypt(modified_plaintext_int)
        
        # Calculate avalanche effect
        avalanche_effect = calculate_avalanche_effect(ciphertext, modified_ciphertext)
        
        # Store the results in the dictionary
        results[plaintext]["Modified_Plaintext"] = modified_plaintext
        results[plaintext]["Key"] = key_int
        results[plaintext][f"Simon_{mode}_Avalanche"] = avalanche_effect

# Create a pandas DataFrame from the results
df = pd.DataFrame.from_dict(results, orient="index")

# Save the DataFrame to a CSV file
df.to_csv("encryption_results.csv", index_label="Plaintext")
```
